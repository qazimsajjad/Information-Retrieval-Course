```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def imshow(image, title=None):
    plt.axis('off')
    plt.title(title)
    plt.imshow(image, 'gray')
    plt.show()
```

```python
query_img = cv.imread('ball1.jpg')
train_img = cv.imread('ball2.jpg')
```

```python
# Convert it to grayscale 
query_img_g = cv.cvtColor(query_img, cv.COLOR_BGR2GRAY) 
train_img_g = cv.cvtColor(train_img, cv.COLOR_BGR2GRAY) 
```

```python
# Initialize the orb detector algorithm
orb = cv.ORB_create()
```

```python
# Now detect the keypoints and compute the descriptors for the query image and train image 
queryKeypoints, queryDescriptors = orb.detectAndCompute(query_img_g, None) 
trainKeypoints, trainDescriptors = orb.detectAndCompute(train_img_g, None) 
```

```python
# Initialize the Matcher for matching the keypoints and then match the keypoints 
matcher = cv.BFMatcher()
matches = matcher.match(queryDescriptors, trainDescriptors)
```

```python
# Use drawMatches() to display matched keypoints between query and train images
final_img = cv.drawMatches(query_img, queryKeypoints,train_img, trainKeypoints, matches[:15],None)
```

```python
final_img = cv.resize(final_img, (1000, 650))
```

```python
imshow(final_img, 'Output')
```

```python
cv.imshow("Output", final_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
